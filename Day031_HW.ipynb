### 用上述網站說明該如何判斷 Header 中必須加上哪些資料才會正確


```python
import requests
url = 'https://www.ptt.cc/bbs/NBA/index.html'
header = {
'accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.9',
'accept-encoding': 'gzip, deflate',
'accept-language': 'zh-TW,zh;q=0.9,en-US;q=0.8,en;q=0.7',
'cache-control': 'max-age=0',
'cookie': '_ga=GA1.2.1366666447.1588560189; __cfduid=d344472549cc654110d800af58cf65c851595566726; __cf_bm=11bff306a3c47e66207664754e9ab215ed97b19f-1597905324-1800-AWYavO6z51jnWL17Fw8WdQR7JUSWd7OLdSMqFVWlhEXQEZdK4SeVWGt7zqB0wnYt9c8r2oPQ7d7xUgt/cSZSS/w=; _gid=GA1.2.164386683.1597905325; _gat=1',
'referer': 'https://www.ptt.cc/bbs/index.html',
'sec-fetch-dest': 'document',
'sec-fetch-mode': 'navigate',
'sec-fetch-site': 'same-origin',
'sec-fetch-user': '?1',
'upgrade-insecure-requests': '1',
'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/84.0.4147.125 Safari/537.36',
}
r = requests.get(url, headers=header)
r.text
```
