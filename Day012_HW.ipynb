### ① 取出今天所有的發文

```python
import requests
from bs4 import BeautifulSoup
url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html.parser")

for d in soup.find(class_='part_list_2').find_all('h3'):
    print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
cls={}
dic = {}
li = []
for d in soup.find(class_="part_list_2").find_all('h3'): 
    ta= d.find(class_='tag').text
    ti= d.find(class_='date').text + d.find_all('a')[-1].text
    if( ta in cls):
        cls[ta]=cls[ta]+1
        dic[ta].append(ti)
    else:
        cls[ta]=0
        dic[ta]=[ti]
print(dic)
```

### ③ 哪一個類別的文章最多

```python
print('文章最多的類別', max(cls, key= cls.get), max(cls.values()))
```
